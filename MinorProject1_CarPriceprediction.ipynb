```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.linear_model import Lasso
from sklearn import metrics
```

```python
#checkin count of fuel type (categorical)
print(df.Fuel_Type.value_counts())
print(df.Seller_Type.value_counts())
print(df.Transmission.value_counts())
```

```python
#label encoding the categorical data
df.replace({"Fuel_Type":{"Petrol":0, "Diesel":1, "CNG":2}}, inplace=True)

df.replace({"Seller_Type":{"Dealer":0, "Individual":1}}, inplace=True)

df.replace({"Transmission":{"Manual":0, "Automatic":1}}, inplace=True)
```

```python
df.head()
```

```python
#splitting data
X=df.drop(["Car_Name", "Selling_Price"], axis=1)
Y=df["Selling_Price"]
```

```python
print(X)
```

```python
print(Y)
```

```python
X_train, X_test, Y_train, Y_test=train_test_split(X,Y, test_size=0.2, random_state=3)
```

Linear Regression

```python
linear_regression=LinearRegression()

linear_regression.fit(X_train, Y_train)
```

```python
#model eval
data_pred=linear_regression.predict(X_train)
```

```python
#visualisation
plt.scatter(Y_train, data_pred)
plt.xlabel("Actual Price")
plt.ylabel("Predicted Price")
plt.title("Actual vs Predicted Prices")
plt.show()
```

```python
data_pred=linear_regression.predict(X_test)
```

```python
#visualisation
plt.scatter(Y_test, data_pred)
plt.xlabel("Actual Price")
plt.ylabel("Predicted Price")
plt.title("Actual vs Predicted Prices")
plt.show()
```

Lasso Regression

```python
lasso_regression=Lasso()

lasso_regression.fit(X_train, Y_train)
```

```python
#model eval
data_pred=lasso_regression.predict(X_train)
```

```python
error_score=metrics.r2_score(Y_train, data_pred)
print("R squared error: ", error_score)
```

```python
#visualisation
plt.scatter(Y_train, data_pred)
plt.xlabel("Actual Price")
plt.ylabel("Predicted Price")
plt.title("Actual vs Predicted Prices")
plt.show()
```
